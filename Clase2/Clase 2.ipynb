```python
%matplotlib notebook
# %matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
#Plot en 3D
from mpl_toolkits.mplot3d import axes3d
```

```python
fig, ax = plt.subplots()

x = np.arange(0.,2.*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))
```

```python
def Animate(i):
    line.set_ydata(np.sin(x - i / 10))
    return line,

TMax = 50
dt = 0.01
NSteps = 100
frames = int(TMax / float(NSteps * dt))
```

```python
Animation = anim.FuncAnimation(fig, Animate, frames=frames, interval=10, blit=True)
```

```python
fig3 = plt.figure(figsize=(5,5))
ax = fig3.add_subplot(1,1,1, projection='3d')

ax.set_xlim(-1.,1.)
ax.set_ylim(-1.,1.)
ax.set_zlim(-1.,1.)

t= np.linspace(-2*np.pi, 2*np.pi, 50)
x, y, z = np.cos(t), np.sin(t), t/t.max()

ax.view_init(10, 45)

data = np.array([[x,y,z],[x,y,z]])

lines = [ax.plot(dat[0, 0:1], dat[1, 0:1], dat[2, 0:1])[0] for dat in data]

for dat, line in zip(data, lines):
    line.set_data(dat[0:2, :50])
    line.set_3d_properties(dat[2, :50])
    line.set_marker('.')
    line.set_color('b')
```

```python
#Caminata aleatoria en 3D


def Update(num, dataLines, lines):
    for line, data in zip(lines, dataLines):
        # primeras dos filas, numero de columnas para la animacion
        line.set_data(data[0:2, :num])
        line.set_3d_properties(data[2, :num])
        line.set_marker('.')
        line.set_color('r')
        dataLines = []
    return lines
```
