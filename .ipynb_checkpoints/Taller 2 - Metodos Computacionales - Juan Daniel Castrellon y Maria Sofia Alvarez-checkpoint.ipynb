<h1>1. Método de Newton-Raphson</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return (3*x**5 + 5*x**4 - x**3)
```

```python
def derivative(f, x, h):
    d = 0.0
    if (h!=0):
        d = (f(x+h) - f(x-h))/(2.0*h)
    return d
```

```python
xi, xf, Npoints = -2., 1., 50
h = (xf-xi)/float(Npoints)
```

```python
X = np.linspace(xi,xf,Npoints)
Y = function(X)
```

```python
plt.plot(X,Y,'b')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.grid()
plt.show()
```

```python
def NewtonMethod(f, df, iteraciones, range_initial, range_end, points, precision, tolerancia):
    possiblePoints = np.linspace(range_initial, range_end, points)
    
    raices = []
    
    for i in possiblePoints:
        xn = i
        j = 0
        error = 100
        while ((error > precision) and (j < iteraciones)):
            xn1 = xn - (f(xn))/df(f,xn,h)
            error = abs(xn1-xn)/abs(xn1)
            xn = xn1
            j += 1
        yn = f(xn)
        anadir = True
        for k in raices:
            if(xn < k + tolerancia and xn > k - tolerancia):
                anadir = False
                break
        if anadir:
            raices.append(xn)
        
    return raices
```

```python
raices = NewtonMethod(function, derivative, 2000, -4, 4, 100, 0.001, 0.01) 
```

```python
plt.plot(X,Y,'b')
for i in raices:
    plt.plot(i, 0, 'ro')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.grid()
plt.show()
```
